```
import numpy as np

m=2;n=4;alpha=0.95;

p0=np.array([
              [0.1, 0.3, 0.6, 0.0],
              [0.0, 0.2, 0.5, 0.3],
              [0.1, 0.1, 0.2, 0.7],
              [0.8, 0.1, 0.0, 0.1]
            ])
p1=np.array([
              [0.6, 0.3, 0.1, 0.0],
              [0.75, 0.1, 0.1, 0.05],
              [0.8, 0.2, 0.0, 0.0],
              [0.9, 0.1, 0.0, 0.0]
            ])
p=[p0,p1]


f0=np.array([100,125,150,500])
f1=np.array([300, 325, 350, 600])
f=[f0,f1]

v=np.zeros(n)
delta=100; epsilon=0.0001;

while (delta>epsilon):
    q=np.zeros((m,n))
    for i in range(m):
          q[i]=f[i]+alpha*p[i].dot(v)
    w=q[0]
    
    for i in range(n):
        for j in range(m):
            if q[j][i]<w[i]:
                w[i]=q[j][i]
    delta=max(abs(v-w))
    v=w
    print(v)
    


for i in range(m):
    q[i]=f[i]+alpha*p[i].dot(v)
w=q[0]
a=np.zeros(n,dtype=np.int64)
for i in range(n):
     for j in range(m):
          if q[j][i]<w[i]:
              w[i]=q[j][i]
              a[i]=j
print(a)
    


```
